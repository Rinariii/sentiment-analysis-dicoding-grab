#Install dan Import Library yang diperlukan

```python
from google_play_scraper import app, reviews
import pandas as pd
```

#Lakukan Scraping Data dari Google PlayStore

```python
# Definisikan ID aplikasi yang akan di review di playstore
app = 'com.grabtaxi.passenger'

# Mengambil 10000 review
data, _ = reviews(
    app,
    lang='id',
    country='id',
    count=10000
)

# Mengubah hasil menjadi dataframe
df = pd.DataFrame([{
    'id': r['reviewId'],
    'nama': r['userName'],
    'rating': r['score'],
    'tanggal': r['at'],
    'Review': r['content']
} for r in data]).drop_duplicates(subset='id') # Jika ada ID yang duplikat maka akan dihapus

# Menyimpan file ke CSV untuk analisis lanjutan
df.to_csv('grab_review.csv', index=False, encoding='utf-8-sig')

# Cek hasil CSV
df
```
